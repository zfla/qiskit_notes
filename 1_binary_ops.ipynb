```python
from qiskit import QuantumCircuit, assemble, Aer
from qiskit.visualization import plot_histogram
```

### Creating a quantum circuit 
This is a circuit containing 8 qubits, and 8 outputs.

```python
qc_output = QuantumCircuit(8)
```

Here, an output is extracted, with each measurement telling a specific qubit to yield an output to an output bit.

```python
qc_output.measure_all()
```

```python
qc_output.draw(initial_state=True)
```

The probability of reaching this output is 100% since there are no quantum effects


```python
sim = Aer.get_backend('aer_simulator')
result = sim.run(qc_output).result()
counts = result.get_counts()
plot_histogram(counts)
```

### Creating an adding circuit
First, we encode inputs with a NOT gate (x), before extracting results using the qc_output circuit.

```python
qc_encode = QuantumCircuit(8)
qc_encode.x(7)
qc_encode.draw()
```

```python
qc_encode.measure_all()
qc_encode.draw()
```

```python
sim = Aer.get_backend("aer_simulator")
result = sim.run(qc_encode).result()
counts = result.get_counts()
plot_histogram(counts)
```

We notice from this that the NOT gate manipulated the final qubit (on the left), leaving the others. Following on from this, we can explore the use of a XOR gate to represent binary addition. The XOR (CNOT) gate has a control (one with the dot) and the target (circle with a plus) qubit. This gate works by checking to see if both bits are equal. If they are the same, the target is changed to equal 0, and is changed to a 1 otherwise.

```python
qc_cnot = QuantumCircuit(2)
qc_cnot.cx(0, 1)
qc_cnot.draw()
```

For the half-adder that we will create, we will write the result onto two other qubits instead of overwriting one. Two CNOTs are used for this. A Toffoli gate will be used to check if both qubits 0 and 1 are equal to 1, to write 1 to qubit 3.

```python
qc_ha = QuantumCircuit(4,2)
# encoding inputs into qubits 0 and 1 (using a NOT to set both qubits to 1)
qc_ha.x(0)
qc_ha.x(1)
qc_ha.barrier()
# using cnots to write XOR of the inputs onto qubit 2
qc_ha.cx(0, 2)
qc_ha.cx(1, 2)
qc_ha.ccx(0, 1, 3)
qc_ha.barrier()
# extracting outputs
qc_ha.measure(2, 0)
qc_ha.measure(3, 1)
qc_ha.draw()
```

```python
qobj = assemble(qc_ha)
counts = sim.run(qobj).result().get_counts()
plot_histogram(counts)
```

We expect the output to be 10. 1 + 1 = 10 in base 2.
